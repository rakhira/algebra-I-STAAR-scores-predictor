```python
counter_1 = 0
for i in df.School_Num:
    url = f'https://rptsvr1.tea.texas.gov/cgi/sas/broker?_service=marykay&_debug=0&single=N&batch=N&app=PUBLIC&ptype=H&_program=perfrept.perfmast.sas&level=campus&search=campnum&namenum={i}&prgopt=2019/tapr/performance.sas'
    r = requests.get(url)
    soup1 = BeautifulSoup(r.content, 'html.parser')
    time.sleep(.5)
    client = MongoClient()
    pages = db.staar
    pages.insert_one({'html_staar': r.content, 'staar_school_id': i})
    counter_1 += 1
```

```python
counter_2 = 0
for i in df.School_Num:
    url = f'https://rptsvr1.tea.texas.gov/cgi/sas/broker?_service=marykay&_debug=0&single=N&batch=N&app=PUBLIC&ptype=H&_program=perfrept.perfmast.sas&level=campus&search=campnum&namenum={i}&prgopt=2019/tapr/staff.sas'
    r = requests.get(url)
    soup2 = BeautifulSoup(r.content, 'html.parser')
    time.sleep(.5)
    client = MongoClient()
    pages = db.staff
    pages.insert_one({'html_staff': r.content, 'staff_school_id': i})
    counter_2 += 1
```
